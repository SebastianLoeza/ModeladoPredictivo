# Regresión Logística

## Importar bibliotecas y/o librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar dataset y cargar las VI y VD

```python
dataset = pd.read_csv('Social_Network_Ads.csv')


X = dataset.iloc[: , 2:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

## Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba =100 observaciones

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
##Imprimir datos de entrenamiento: vi y vd
print(X_test)
print(X_train)
```

## Escalado de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Entrenamiento del modelo de regresión logística en el conjunto de entrenamiento
###**revisar la documentación de LogisticRegression para conocer los principales hiperparámetros

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicción de un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicción de los resultados del conjunto de pruebas

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Realizar la matriz de confiusión
### para ello usaran la siguiente función "confusion_matrix"

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay


cm_iris = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_iris)
disp.plot()
```

## Visualización de los resultados del conjunto de entrenamiento

```python
 from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```
