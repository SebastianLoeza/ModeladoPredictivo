## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model  import LinearRegression

import time
```

## Importing the dataset

```python
dataset = pd.read_csv('position_salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
X
```

### Create the Polynomial Models

```python
#Regresión Polinomial de 2ndo Grado

inicio = time.time()

poly_reg_2 = PolynomialFeatures(degree=2)
X_poly_2= poly_reg_2.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly_2,y)

fin = time.time()
tiempo_transcurrido = fin - inicio
print("El código tardó:", tiempo_transcurrido, "segundos")
```

```python
#Regresión Plinomial de 3er grado

inicio = time.time()

poly_reg_3 = PolynomialFeatures(degree=3)
X_poly_3= poly_reg_3.fit_transform(X)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_3, y)

fin = time.time()
tiempo_transcurrido = fin - inicio
print("El código tardó:", tiempo_transcurrido, "segundos")
```

```python
#Regresión Polinomial de 4to grado

inicio = time.time()

poly_reg_4 = PolynomialFeatures(degree=4)
X_poly_4 = poly_reg_4.fit_transform(X)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly_4, y)

fin = time.time()
tiempo_transcurrido = fin - inicio
print("El código tardó:", tiempo_transcurrido, "segundos")
```

```python
#Regresión Polinomial de 8vo grado
inicio = time.time()

poly_reg_8 = PolynomialFeatures(degree=8)
X_poly_8 = poly_reg_8.fit_transform(X)
lin_reg_8 = LinearRegression()
lin_reg_8.fit(X_poly_8, y)

fin = time.time()
tiempo_transcurrido = fin - inicio
print("El código tardó:", tiempo_transcurrido, "segundos")
```

## Plot the test set results of the polynomial regressor's

```python
plt.scatter(X,y, color = 'red')
plt.plot(X,lin_reg_2.predict(poly_reg_2.fit_transform(X)), color = 'blue')
plt.title('Nivel vs  Salario (Regresión Polinomial de 2ndo grado)')
plt.xlabel('Nivel')
plt.ylabel('Salario')
plt.show()
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X,lin_reg_3.predict(poly_reg_3.fit_transform(X)), color = 'blue')
plt.title('Nivel vs  Salario (Regresión Polinomial de 3er grado)')
plt.xlabel('Nivel')
plt.ylabel('Salario')
plt.show()
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X,lin_reg_4.predict(poly_reg_4.fit_transform(X)),color ='blue')
plt.title('Nivel vs  Salario (Regresión Polinomial de 4to grado)')
plt.xlabel('Nivel')
plt.ylabel('Salario')
plt.show()
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X,lin_reg_8.predict(poly_reg_8.fit_transform(X)),color ='blue')
plt.title('Nivel vs  Salario (Regresión Polinomial de 8vo grado)')
plt.xlabel('Nivel')
plt.ylabel('Salario')
plt.show()
```

## Predicting the Test set results

```python
y_pred_2 = lin_reg_2.predict(poly_reg_2.fit_transform([[6.5]]))
y_pred_3 = lin_reg_3.predict(poly_reg_3.fit_transform([[6.5]]))
y_pred_4 = lin_reg_4.predict(poly_reg_4.fit_transform([[6.5]]))
```

```python
print('Prediccion de 6.5 con Regresión Polinomial de 2ndo grado: ',y_pred_2[0])
print('Prediccion de 6.5 con Regresión Polinomial de 3er grado: ',y_pred_3[0])
print('Prediccion de 6.5 con Regresión Polinomial de 4to grado: ',y_pred_4[0])
```
